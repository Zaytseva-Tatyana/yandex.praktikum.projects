## Подготовка данных

```python
# приведение названий столбцов к нижнему регистру
data.columns = data.columns.str.lower()
```

```python
# удаление из набора данных срок с пропущенными значениями в name или year_of_release
data.dropna(subset=['name', 'year_of_release'], inplace=True)
data.info()
```

```python
data['critic_score'] = data['critic_score'] / 10
data
```

```python
data['user_score'] = data['user_score'].replace('tbd', None)
data.info()
```

```python

def rating(row):
    if row['rating'] == 'E' or row['rating'] == 'K-A':
        return 'Для всех'
    if row['rating'] == 'M':
        return 'Для взрослых'
    if row['rating'] == 'T':
        return 'Для подростков'
    if row['rating'] == 'E10+':
        return 'Для всех от 10 лет и старше'
    if row['rating'] == 'AO':
        return 'Только для взрослых'
    if row['rating'] == 'EC':
        return 'Для детей младшего возраста'
    if row['rating'] == 'RP':
        return 'Рейтинг ожидается'
    else:
        return 
    
data['rating_rus'] = data.apply(rating, axis=1)
```

```python
data
```

```python
data['year_of_release'] = data['year_of_release'].astype('int')
data['user_score'] = data['user_score'].astype('float')
```

```python
data['sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data
```

# Проверка гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"

**H0:** Средние пользовательские рейтинги платформ Xbox One и PC равны

**H1:** Средние пользовательские рейтинги платформ Xbox One и PC различаются


```python
from scipy import stats as st
import numpy as np

ps4_user_score = actual_data.query('platform=="PS4" and user_score.isna()==False')['user_score']

xone_user_score = actual_data.query('platform=="XOne" and user_score.isna()==False')['user_score']

ps4_var = np.var(ps4_user_score)
xone_var = np.var(xone_user_score)

print('Дисперсия пользовательского рейтинга платформы PS4', ps4_var)
print('Дисперсия пользовательского рейтинга платформы Xone', xone_var)

```

```python
alpha = .05

results = st.ttest_ind(ps4_user_score, xone_user_score, equal_var=True)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

# Вывод

Средние пользовательские рейтинги платформ Xbox One и PC равны.

# Проверка гипотезы "Средние пользовательские рейтинги жанров Action и Sports разные"

**H0:** Средние пользовательские рейтинги жанров Action и Sports равны.

**H1:** Средние пользовательские рейтинги жанров Action и Sports разные.

```python
action_user_score = actual_data.query('genre=="Action" and user_score.isna()==False')['user_score']

sports_user_score = actual_data.query('genre=="Sports" and user_score.isna()==False')['user_score']

action_var = np.var(action_user_score)
sports_var = np.var(sports_user_score)

print('Дисперсия пользовательского рейтинга жанра Action', action_var)
print('Дисперсия пользовательского рейтинга жанра Sports', sports_var)
```

```python
alpha = .05

results = st.ttest_ind(action_user_score, sports_user_score, equal_var=True)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам статистического теста нулевая гипотеза отвергнута. А значит, альтернативная гипотеза "Средние пользовательские рейтинги жанров Action и Sports разные" верна.
