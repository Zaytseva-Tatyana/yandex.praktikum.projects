### Шаг 2. Предобработка данных

```python
data[data['balcony'].isna()==True]
```

Наличие балкона влияет на стоимость квартиры, поэтому будем считать, что если пользователь не указал в объявлении кол-во балконов, значит их нет.

```python
data['balcony'] = data['balcony'].fillna(0).astype('int')
data['balcony'].unique()
```

Почти у половины набора данных не указана высота потолков. Т.к это значение может быть важным для формирования цены, его невозможно заполнить, не искажая статистику, поэтому отсутствующие значения будут заменены на 0.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(0)
data['ceiling_height'].value_counts().sort_values()
```

```python
# приведение даты к типу datetime
data['first_day_exposition'] =  pd.to_datetime(data['first_day_exposition'])
```

### Шаг 3. Обогащение данных

```python
# добавление колонки с ценой за квадратный метр
data['price_per_meter'] = data['last_price'] / data['total_area'] 
# добавление колонки в какой день недели опубликовано объявление
data['weekday_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).weekday 
# добавление колонки с месяцем публикации объявления
data['month_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).month
# добавление колонки с годом публикации объявления
data['year_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).year

# определение функции, которая категоризует квартиры в зависимости от этажа: первый, последний и другой
def floor_category(data):
    if data['floor'] == 1:
        return 'первый'
    if data['floor'] == data['floors_total']:
        return 'последний'
    else:
        return 'другой'

# добавление колонки с категорией этажа
data['name_floor'] = data.apply(floor_category, axis=1)
# добавление колонки с долей жилой площади от общей
data['living_ratio'] = data['living_area'] / data['total_area']
# добавление колонки с долей площади кухни от общей
data['kitchen_ratio'] = data['kitchen_area'] / data['total_area']
display(data)
```

Пропущенные значения на основе картографических данных будут оставлены как есть, т.к. их невозможно восстановить.

В 2015 году было резкое падение цен на недвижимость по всему городу. С 2015 по 2018 цена на недвижимость в центре оставалась стабильной, в то время как в остальных частях города продолжала снижаться. В 2018 квартиры в центре города подешевели, хотя по всему Петербургу наблюдается повышение цены.

Анализ выявил общие факторы ценообразования:
- населенный пункт
- удаленность от центра
- количество комнат
- общая площадь квартиры
- этаж
- высота потолков
